To verify the power of dreams, we must first break the brain.
We will **temporarily disable** its ability to merge memories instantly (by raising `resonance_threshold` to 10.0).
This simulates **Sleep Deprivation**: The brain is forced to hold onto every piece of noisy raw data without processing it.


```python
# Create Chaos / åˆ¶é€ æ··ä¹±
print("Generating noise...")
# 1. Temporarily disable 'Automatic Merging' to simulate Sleep Deprivation
#    (If we don't do this, the smart brain will merge them instantly!)
brain.right_hemisphere.resonance_threshold = 10.0

# 2. Feed messy data
concepts = [np.random.rand(784) for _ in range(3)]
names = ['Concept_A', 'Concept_B', 'Concept_C']

for i in range(200): # 600 stars total
    for base, name in zip(concepts, names):
        # Low variance (0.05) ensures they are close enough to eventually merge
        noise = np.random.normal(0, 0.05, 784)
        brain.memorize(base + noise, name)

print(f"Stars: {len(brain.galaxy)} (High Entropy)")

# Show the Messy Nebula / å±•ç¤ºæ··ä¹±çš„æ˜Ÿäº‘
fig, msg = get_star_map_figure(brain)
if fig:
    fig.show()
else:
    print(msg)
```
