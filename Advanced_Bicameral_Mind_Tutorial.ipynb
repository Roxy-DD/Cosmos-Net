```python
# 1. Awakening the Twin Gods (Âî§ÈÜíÂèåÂ≠êÁ•û)
# Let's install and import our digital brain.
!git clone https://github.com/Roxy-DD/Cosmos-Net.git 2>/dev/null
%cd Cosmos-Net
!pip install -r requirements.txt -q

from cosmos_net import CorpusCallosum, CosmosPhysics
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

# Wake up!
brain = CorpusCallosum()
print(f"üß† Brain Awakened. Current Boss: {'Intuition (Right)' if brain.dominance > 0.5 else 'Logic (Left)'}")
```

### üî¨ Step 1: Meet the Librarian (Left Hemisphere)
**Biology**: The Left Brain handles **Logic & Language**. It likes distinct categories.
**Code**: `LeftHemisphere` uses **Statistics**. It converts an image into numbers like "Pixel Density" (How much ink?) and "Aspect Ratio" (Is it tall or fat?).

Let's show it a "1" and a "0".

```python
# Load real data
X, y = fetch_openml('mnist_784', version=1, return_X_y=True, as_frame=False, parser='auto')
X = X / 255.0
vec_1 = X[np.where(y == '1')[0][0]]
vec_0 = X[np.where(y == '0')[0][0]]

# Show the Librarian
brain.left_hemisphere.memorize(vec_1, '1')
brain.left_hemisphere.memorize(vec_0, '0')

# Ask it what it thinks
# It calculates the 'Z-Score' (How weird is this?). Low score = High Confidence.
label, conf = brain.left_hemisphere.perceive(vec_1)
print(f"ü§ì Librarian says: 'Based on my measurements, this is a {label}. I am {conf*100}% sure.'")
```

### üé® Step 2: Meet the Artist (Right Hemisphere)
**Biology**: The Right Brain handles **Spatial & Holistic** processing. It recognizes faces and patterns instantly.
**Code**: `RightHemisphere` uses **Gravity (Dot Product)**. It feels the "pull" of similar memories.

The Artist doesn't measure aspect ratios. It just feels the **Vibe**.

```python
# Show the Artist
brain.right_hemisphere.memorize(vec_1, '1')

# Ask it to feel
# It calculates Gravity (0.0 to 1.0)
star, gravity = brain.right_hemisphere.perceive(vec_1)
print(f"üé® Artist says: 'This feels exactly like a {star.label}! The resonance is {gravity:.4f}!'")
```

### ‚öîÔ∏è Step 3: The Argument (Neuroplasticity)
**Biology**: Sometimes, our brain conflicts. Logic names it one thing, Intuition feels another. The **Corpus Callosum** must inhibit one to let the other speak.
**Code**: `brain.dominance`. This number moves Left or Right based on who gets the right answer.

Let's confuse the brain. We will show it a '7', but tell it it's a '1'. (Bad teaching!).

```python
vec_7 = X[np.where(y == '7')[0][0]]

# 1. The Conflict
# Librarian sees a slanted line (maybe a 1?). Artist sees something new.
res, conf = brain.perceive(vec_7)
print(f"ü§î Brain Initial Thought: {res.label}")

# 2. The Environment Corrects
# We tell it: "No! This is a 7!"
msg = brain.memorize(vec_7, '7')
print(f"Outcome: {msg}")

# 3. The Change (Plasticity)
# Look closely at the 'dominance' value. The brain has physically shifted its personality.
print(f"Personality Shifted: Dominance is now {brain.dominance:.2f}")
```

### üí§ Step 4: Time for Bed (The Dreamtime)
**Biology**: We don't remember everything. During sleep (**Rapid Eye Movement**), our brain replays memories. It deletes the weak ones (**Pruning**) and merges specific events into general wisdom (**Consolidation**).
**Code**: `brain.dream()`. 

Let's overload the Artist with 50 messy sketches of '1'.

```python
# A busy day at school...
print(f"Stars before: {len(brain.galaxy)}")
for i in range(50):
    noise = np.random.normal(0, 0.15, vec_1.shape)
    brain.memorize(vec_1 + noise, '1')
print(f"Stars after a long day: {len(brain.galaxy)} (So messy!)")

# Goodnight, Cosmos.
print("üí§ Dreaming...")
report = brain.dream()
print(report)

# Look! The count dropped. The Artist combined all those messy sketches into one Masterpiece.
```
