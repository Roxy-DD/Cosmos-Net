## **ç¬¬ä¸€æ­¥ï¼šå”¤é†’å®‡å®™ (The Magic Spell)**

æˆ‘ä»¬ä¸éœ€è¦æ‡‚å¤æ‚çš„æ•°å­¦ï¼Œä½ åªéœ€è¦çŸ¥é“è¿™ä¸ªå®‡å®™åªæœ‰ä¸¤æ¡ **å®¶è§„**ï¼š

1.  **è´ªåƒè§„åˆ™ (æ³•åˆ™ä¸€)ï¼š** é‡åˆ° **è§è¿‡** çš„ä¸œè¥¿ï¼ŒçŽ°æœ‰çš„æ˜Ÿæ˜Ÿä¼šæŠŠå®ƒâ€œåƒæŽ‰â€ï¼Œå˜å¾—æ›´å¤§æ›´äº®ã€‚
2.  **çˆ†ç‚¸è§„åˆ™ (æ³•åˆ™äºŒ)ï¼š** é‡åˆ° **æ²¡è§è¿‡** çš„ä¸œè¥¿ï¼Œå®‡å®™ä¼šâ€œè½°â€çš„ä¸€å£°ï¼Œè¯žç”Ÿä¸€é¢—å…¨æ–°çš„æ˜Ÿæ˜Ÿï¼

ðŸ‘‡ **ç‚¹å‡»ä¸‹æ–¹çš„å°ä¸‰è§’ï¼Œå¯åŠ¨å¼•æ“Žï¼š**


```python
# ==========================================
# ðŸ”® ä½ çš„å®‡å®™é­”æ³•å¼•æ“Ž (ä¸ç”¨çœ‹æ‡‚ï¼Œè¿è¡Œå®ƒå°±è¡Œï¼)
# ==========================================
import numpy as np
import matplotlib.pyplot as plt

# è¿™æ˜¯"æ˜Ÿæ˜Ÿ"ï¼šæ¯ä¸€ä¸ªè®°å¿†éƒ½æ˜¯ä¸€é¢—å‘å…‰çš„æ˜Ÿ
class Star:
    def __init__(self, name, soul):
        self.name = name       # åå­— (æ¯”å¦‚"è‹¹æžœ")
        self.soul = soul       # çµé­‚ (ä¸€ä¸²ç¥žå¥‡çš„æ•°å­—)
        self.size = 1          # å¤§å° (åƒå¾—è¶Šå¤šï¼Œé•¿å¾—è¶Šå¤§)
        self.color = np.random.rand(3,) # ç»™æ¯é¢—æ˜Ÿä¸€ä¸ªéšæœºé¢œè‰²

# è¿™æ˜¯"å®‡å®™"ï¼šå®ƒè´Ÿè´£çœ‹å’Œå­¦
class Universe:
    def __init__(self):
        self.stars = []        # çŽ°åœ¨çš„æ˜Ÿç³»

    # åŠ¨ä½œï¼šçœ‹ä¸€çœ‹
    def look(self, thing):
        thing = thing / np.linalg.norm(thing)
        best_star = None
        max_love = -1.0
        
        for star in self.stars:
            love = np.dot(thing, star.soul) # ç›¸ä¼¼åº¦
            if love > max_love:
                max_love = love
                best_star = star
        return best_star, max_love

    # åŠ¨ä½œï¼šå­¦ï¼
    def teach(self, name, thing):
        thing = thing / np.linalg.norm(thing)
        best_star, love = self.look(thing)
        
        # å¦‚æžœçˆ±æ„å¾ˆå¼º (>0.9)ï¼Œè¯´æ˜Žè®¤è¯†ï¼
        if best_star and love > 0.9:
            print(f"ðŸ˜‹ [Gluttony]: Ah! I know this [{best_star.name}]! (Gulp! Ate it)")
            best_star.size += 500 # å˜å¤§ä¸€ç‚¹æ–¹ä¾¿çœ‹
            best_star.soul = best_star.soul * 0.9 + thing * 0.1
            best_star.soul /= np.linalg.norm(best_star.soul)
            
        # å¦‚æžœçˆ±æ„å¾ˆå¼±ï¼Œè¯´æ˜Žæ˜¯æ–°æœ‹å‹ï¼
        else:
            print(f"ðŸ’¥ [Big Bang]: Wow! Something brand new! The Universe created [{name}] Star!")
            new_star = Star(name, thing)
            new_star.size = 200 # åˆå§‹å¤§å°
            self.stars.append(new_star)

# ç”»å›¾å°åŠ©æ‰‹
def plot_nebula(universe, title="Cosmos Nebula"):
    plt.figure(figsize=(8, 6))
    plt.title(title, fontsize=15)
    
    # å› ä¸ºå±å¹•åªæœ‰2Dï¼Œæˆ‘ä»¬åªç”»å‡ºæ˜Ÿæ˜Ÿçš„å‰ä¸¤ä¸ªç‰¹å¾ï¼ˆæŠ•å½±ï¼‰
    for star in universe.stars:
        x, y = star.soul[0], star.soul[1]
        plt.scatter(x, y, s=star.size, c=[star.color], alpha=0.6, edgecolors='white', linewidth=2)
        # ç”¨è‹±æ–‡æ ‡ç­¾ï¼Œé¿å…æ–¹å—ä¹±ç 
        label_map = {"è‹¹æžœ": "Apple", "å¤–æ˜Ÿäºº": "Alien"} 
        label_name = label_map.get(star.name, star.name)
        plt.text(x, y+0.05, label_name, fontsize=12, ha='center', color='darkblue')

    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.show()

print("âœ… é­”æ³•å¼•æ“Žå¯åŠ¨æˆåŠŸï¼ä½ çš„å®‡å®™çŽ°åœ¨ä¸€ç‰‡æ¼†é»‘ï¼Œç­‰å¾…ä½ æ¥ç‚¹äº®ã€‚")
```

## **ç¬¬äºŒæ­¥ï¼šå–‚å…»ä½ çš„å®‡å®™ (Feeding Time)**

çŽ°åœ¨å®‡å®™æ˜¯æ´»çš„ï¼Œä½†å®ƒæ˜¯ç©ºçš„ã€‚æˆ‘ä»¬è¦å–‚å®ƒåƒâ€œæ¦‚å¿µâ€ã€‚
ç”µè„‘çœ‹ä¸æ‡‚å›¾ç‰‡ï¼Œæ‰€ä»¥æˆ‘ä»¬ç”¨ **ä¸€ä¸²æ•°å­—** æ¥ä»£è¡¨ä¸‡ç‰©çš„çµé­‚ã€‚

ðŸ‘‡ **è®©æˆ‘ä»¬è¯•è¯•çœ‹ï¼š**

1.  å…ˆå–‚å®ƒä¸€ä¸ª **è‹¹æžœ**ã€‚
2.  å†å–‚å®ƒä¸€ä¸ª **è‹¹æžœ**ï¼ˆçœ‹å®ƒä¼šä¸ä¼šè´ªåƒï¼‰ã€‚
3.  çªç„¶å–‚å®ƒä¸€ä¸ª **å¤–æ˜Ÿäºº**ï¼ˆçœ‹å®ƒä¼šä¸ä¼šå“ä¸€è·³ï¼‰ã€‚


```python
# å¯åŠ¨ä½ çš„å°å®‡å®™
my_baby_universe = Universe()

# è¿™ä¸ªå‡½æ•°å¸®æˆ‘ä»¬æŠŠ"åå­—"å˜æˆ"æ•°å­—çµé­‚"
def create_soul(magic_number):
    np.random.seed(magic_number)
    return np.random.normal(0, 1, 50)


# 1. ç¬¬ä¸€æ¬¡æ•™å®ƒ "è‹¹æžœ"
print("--- ðŸŽ åœºæ™¯ 1ï¼šç¬¬ä¸€æ¬¡è§è‹¹æžœ ---")
apple_soul = create_soul(100) # åˆ¶é€ è‹¹æžœçš„çµé­‚
my_baby_universe.teach("è‹¹æžœ", apple_soul)

# 2. å†ç»™å®ƒçœ‹ä¸€ä¸ª "è‹¹æžœ" (ç¨å¾®æœ‰ç‚¹ä¸ä¸€æ ·ï¼Œä¹Ÿè®¸æ˜¯ä¸ªé’è‹¹æžœ)
print("\n--- ðŸ åœºæ™¯ 2ï¼šåˆè§åˆ°äº†è‹¹æžœ ---")
green_apple = apple_soul + np.random.normal(0, 0.2, 50) 
my_baby_universe.teach("è‹¹æžœ", green_apple)

# 3. çªç„¶ç»™å®ƒçœ‹ "å¤–æ˜Ÿäºº"
print("\n--- ðŸ‘½ åœºæ™¯ 3ï¼šçªè¢­ï¼å¤–æ˜Ÿäººæ¥äº† ---")
alien_soul = create_soul(999) # åˆ¶é€ å¤–æ˜Ÿäººçš„çµé­‚
my_baby_universe.teach("å¤–æ˜Ÿäºº", alien_soul)

```

è¯·è¿è¡Œä¸‹é¢çš„ä»£ç ã€‚ä½ ä¼šçœ‹åˆ°æ˜Ÿæ˜Ÿæ¼‚æµ®åœ¨è¿™ä¸ªæ•°å­¦çš„è™šç©ºä¸­ï¼š

## **ç¬¬å››æ­¥ï¼šç»ˆæžæŒ‘æˆ˜â€”â€”è„‘ç­‹æ€¥è½¬å¼¯ (The Paradox)**

è¿™æ˜¯æœ€åŽ‰å®³çš„ä¸€æ­¥ã€‚ç§‘å­¦å®¶å«å®ƒâ€œå“¥å¾·å°”æ‚–è®ºâ€ï¼Œæˆ‘ä»¬å«å®ƒ **â€œè„‘ç­‹æ€¥è½¬å¼¯â€**ã€‚

**å®ƒä¼šç›´æŽ¥é€šè¿‡â€œå¤§çˆ†ç‚¸â€é•¿å‡ºä¸€ä¸ªæ–°è„‘å­æ¥ç†è§£å®ƒï¼**

> âš ï¸ **å›¾è¡¨å°è¯å…¸**ï¼š
> *   **Exploration** = æŽ¢ç´¢æœŸ (åƒå­©å­ä¸€æ ·å¥½å¥‡)
