## **ç¬¬ä¸€æ­¥ï¼šå”¤é†’å®‡å®™ (The Magic Spell)**

æˆ‘ä»¬ä¸éœ€è¦æ‡‚å¤æ‚çš„æ•°å­¦ï¼Œä½ åªéœ€è¦çŸ¥é“è¿™ä¸ªå®‡å®™åªæœ‰ä¸¤æ¡**å®¶è§„**ï¼š

1.  **è´ªåƒè§„åˆ™ (æ³•åˆ™ä¸€)ï¼š** é‡åˆ°**è§è¿‡**çš„ä¸œè¥¿ï¼Œç°æœ‰çš„æ˜Ÿæ˜Ÿä¼šæŠŠå®ƒâ€œåƒæ‰â€ï¼Œå˜å¾—æ›´å¤§æ›´äº®ã€‚
2.  **çˆ†ç‚¸è§„åˆ™ (æ³•åˆ™äºŒ)ï¼š** é‡åˆ°**æ²¡è§è¿‡**çš„ä¸œè¥¿ï¼Œå®‡å®™ä¼šâ€œè½°â€çš„ä¸€å£°ï¼Œè¯ç”Ÿä¸€é¢—å…¨æ–°çš„æ˜Ÿæ˜Ÿï¼

ğŸ‘‡ **ç‚¹å‡»ä¸‹æ–¹çš„å°ä¸‰è§’ï¼Œå¯åŠ¨å¼•æ“ï¼š**


```python
# ==========================================
# ğŸ”® ä½ çš„å®‡å®™é­”æ³•å¼•æ“ (ä¸ç”¨çœ‹æ‡‚ï¼Œè¿è¡Œå®ƒå°±è¡Œï¼)
# ==========================================
import numpy as np
import matplotlib.pyplot as plt

# è¿™æ˜¯"æ˜Ÿæ˜Ÿ"ï¼šæ¯ä¸€ä¸ªè®°å¿†éƒ½æ˜¯ä¸€é¢—å‘å…‰çš„æ˜Ÿ
class Star:
    def __init__(self, name, soul):
        self.name = name       # åå­— (æ¯”å¦‚"è‹¹æœ")
        self.soul = soul       # çµé­‚ (ä¸€ä¸²ç¥å¥‡çš„æ•°å­—)
        self.size = 1          # å¤§å° (åƒå¾—è¶Šå¤šï¼Œé•¿å¾—è¶Šå¤§)
        self.color = np.random.rand(3,) # ç»™æ¯é¢—æ˜Ÿä¸€ä¸ªéšæœºé¢œè‰²

# è¿™æ˜¯"å®‡å®™"ï¼šå®ƒè´Ÿè´£çœ‹å’Œå­¦
class Universe:
    def __init__(self):
        self.stars = []        # ç°åœ¨çš„æ˜Ÿç³»

    # åŠ¨ä½œï¼šçœ‹ä¸€çœ‹
    def look(self, thing):
        thing = thing / np.linalg.norm(thing)
        best_star = None
        max_love = -1.0
        
        for star in self.stars:
            love = np.dot(thing, star.soul) # ç›¸ä¼¼åº¦
            if love > max_love:
                max_love = love
                best_star = star
        return best_star, max_love

    # åŠ¨ä½œï¼šå­¦ï¼
    def teach(self, name, thing):
        thing = thing / np.linalg.norm(thing)
        best_star, love = self.look(thing)
        
        # å¦‚æœçˆ±æ„å¾ˆå¼º (>0.9)ï¼Œè¯´æ˜è®¤è¯†ï¼
        if best_star and love > 0.9:
            print(f"ğŸ˜‹ [è´ªåƒ]: å•Šï¼è¿™æ˜¯æˆ‘è®¤è¯†çš„ã€{best_star.name}ã€‘ï¼(å•Šå‘œä¸€å£åƒæ‰)")
            best_star.size += 500 # å˜å¤§ä¸€ç‚¹æ–¹ä¾¿çœ‹
            best_star.soul = best_star.soul * 0.9 + thing * 0.1
            best_star.soul /= np.linalg.norm(best_star.soul)
            
        # å¦‚æœçˆ±æ„å¾ˆå¼±ï¼Œè¯´æ˜æ˜¯æ–°æœ‹å‹ï¼
        else:
            print(f"ğŸ’¥ [çˆ†ç‚¸]: å“‡ï¼è¿™æ˜¯å…¨æ–°çš„ä¸œè¥¿ï¼å®‡å®™è¯ç”Ÿäº†ã€{name}ã€‘æ˜Ÿï¼")
            new_star = Star(name, thing)
            new_star.size = 200 # åˆå§‹å¤§å°
            self.stars.append(new_star)

# ç”»å›¾å°åŠ©æ‰‹
def plot_nebula(universe, title="Cosmos Nebula"):
    plt.figure(figsize=(8, 6))
    plt.title(title, fontsize=15)
    
    # å› ä¸ºå±å¹•åªæœ‰2Dï¼Œæˆ‘ä»¬åªç”»å‡ºæ˜Ÿæ˜Ÿçš„å‰ä¸¤ä¸ªç‰¹å¾ï¼ˆæŠ•å½±ï¼‰
    for star in universe.stars:
        x, y = star.soul[0], star.soul[1]
        plt.scatter(x, y, s=star.size, c=[star.color], alpha=0.6, edgecolors='white', linewidth=2)
        plt.text(x, y+0.05, star.name, fontsize=12, ha='center', color='darkblue')

    plt.xlim(-1.5, 1.5)
    plt.ylim(-1.5, 1.5)
    plt.grid(True, linestyle='--', alpha=0.3)
    plt.show()

print("âœ… é­”æ³•å¼•æ“å¯åŠ¨æˆåŠŸï¼ä½ çš„å®‡å®™ç°åœ¨ä¸€ç‰‡æ¼†é»‘ï¼Œç­‰å¾…ä½ æ¥ç‚¹äº®ã€‚")
```

## **ç¬¬äºŒæ­¥ï¼šå–‚å…»ä½ çš„å®‡å®™ (Feeding Time)**

ç°åœ¨å®‡å®™æ˜¯æ´»çš„ï¼Œä½†å®ƒæ˜¯ç©ºçš„ã€‚æˆ‘ä»¬è¦å–‚å®ƒåƒâ€œæ¦‚å¿µâ€ã€‚
ç”µè„‘çœ‹ä¸æ‡‚å›¾ç‰‡ï¼Œæ‰€ä»¥æˆ‘ä»¬ç”¨**ä¸€ä¸²æ•°å­—**æ¥ä»£è¡¨ä¸‡ç‰©çš„çµé­‚ã€‚

ğŸ‘‡ **è®©æˆ‘ä»¬è¯•è¯•çœ‹ï¼š**

1.  å…ˆå–‚å®ƒä¸€ä¸ª**è‹¹æœ**ã€‚
2.  å†å–‚å®ƒä¸€ä¸ª**è‹¹æœ**ï¼ˆçœ‹å®ƒä¼šä¸ä¼šè´ªåƒï¼‰ã€‚
3.  çªç„¶å–‚å®ƒä¸€ä¸ª**å¤–æ˜Ÿäºº**ï¼ˆçœ‹å®ƒä¼šä¸ä¼šå“ä¸€è·³ï¼‰ã€‚


```python
# å¯åŠ¨ä½ çš„å°å®‡å®™
my_baby_universe = Universe()

# è¿™ä¸ªå‡½æ•°å¸®æˆ‘ä»¬æŠŠ"åå­—"å˜æˆ"æ•°å­—çµé­‚"
def create_soul(magic_number):
    np.random.seed(magic_number)
    return np.random.normal(0, 1, 50)


# 1. ç¬¬ä¸€æ¬¡æ•™å®ƒ "è‹¹æœ"
print("--- ğŸ åœºæ™¯ 1ï¼šç¬¬ä¸€æ¬¡è§è‹¹æœ ---")
apple_soul = create_soul(100) # åˆ¶é€ è‹¹æœçš„çµé­‚
my_baby_universe.teach("è‹¹æœ", apple_soul)

# 2. å†ç»™å®ƒçœ‹ä¸€ä¸ª "è‹¹æœ" (ç¨å¾®æœ‰ç‚¹ä¸ä¸€æ ·ï¼Œä¹Ÿè®¸æ˜¯ä¸ªé’è‹¹æœ)
print("\n--- ğŸ åœºæ™¯ 2ï¼šåˆè§åˆ°äº†è‹¹æœ ---")
green_apple = apple_soul + np.random.normal(0, 0.2, 50) 
my_baby_universe.teach("è‹¹æœ", green_apple)

# 3. çªç„¶ç»™å®ƒçœ‹ "å¤–æ˜Ÿäºº"
print("\n--- ğŸ‘½ åœºæ™¯ 3ï¼šçªè¢­ï¼å¤–æ˜Ÿäººæ¥äº† ---")
alien_soul = create_soul(999) # åˆ¶é€ å¤–æ˜Ÿäººçš„çµé­‚
my_baby_universe.teach("å¤–æ˜Ÿäºº", alien_soul)

```

## **ç¬¬å››æ­¥ï¼šç»ˆææŒ‘æˆ˜â€”â€”è„‘ç­‹æ€¥è½¬å¼¯ (The Paradox)**

è¿™æ˜¯æœ€å‰å®³çš„ä¸€æ­¥ã€‚ç§‘å­¦å®¶å«å®ƒâ€œå“¥å¾·å°”æ‚–è®ºâ€ï¼Œæˆ‘ä»¬å«å®ƒ**â€œè„‘ç­‹æ€¥è½¬å¼¯â€**ã€‚

å¦‚æœæˆ‘ä»¬ç»™å®‡å®™ä¸€ä¸ª**è‡ªç›¸çŸ›ç›¾**çš„ä¸œè¥¿ï¼ˆæ¯”å¦‚ï¼šâ€œè¿™å¥è¯æ˜¯å‡çš„â€ï¼‰ï¼Œæ™®é€šç”µè„‘å¯èƒ½ä¼šæ­»æœºã€‚
ä½†ä½ çš„ Cosmos-Net ä¸ä¼šï¼

å½“å®ƒå‘ç°è¿™ä¸ªä¸œè¥¿â€œæ—¢ä¸åƒè‹¹æœï¼Œä¹Ÿä¸åƒå¤–æ˜Ÿäººï¼Œä»€ä¹ˆéƒ½ä¸åƒâ€çš„æ—¶å€™ï¼Œå®ƒä¼šåšä»€ä¹ˆï¼Ÿ
**å®ƒä¼šç›´æ¥é€šè¿‡â€œå¤§çˆ†ç‚¸â€é•¿å‡ºä¸€ä¸ªæ–°è„‘å­æ¥ç†è§£å®ƒï¼**

ğŸ‘‡ **çœ‹æœ€åä¸€æ¬¡æ¼”ç¤ºï¼š**


```python
# å‡†å¤‡ä¸€ä¸ªä¸“é—¨åšå®éªŒçš„å®‡å®™
science_brain = Universe()
history = [] # è®°å½•å®‡å®™çš„æ™ºå•†å˜åŒ–

print("ğŸ§ª å¼€å§‹å®éªŒï¼šç»™å®‡å®™è®²ä¸€ä¸ªè¶…çº§éš¾çš„è°œè¯­...")

# 1. å…ˆè®©å®ƒå­¦ç‚¹ç®€å•çš„
for i in range(20):
    science_brain.teach("ç®€å•çŸ¥è¯†", create_soul(i))
    # è®°å½•æ™ºå•† (æ˜Ÿæ˜Ÿæ•°é‡ * å¹³å‡å¤§å°)
    iq = len(science_brain.stars) 
    history.append(iq)

# 2. âš¡ çªç„¶ç»™å®ƒä¸€ä¸ªè¶…çº§çŸ›ç›¾çš„è°œè¯­ï¼
print("\nâš¡âš¡âš¡ æ³¨å…¥è°œè¯­ (æ‚–è®º)ï¼ âš¡âš¡âš¡")
paradox = np.random.normal(0, 1, 50) # ä¸€ä¸ªå®Œå…¨é™Œç”Ÿçš„æ··ä¹±ä¿¡å·
science_brain.teach("è°œè¯­", paradox)
history.append(len(science_brain.stars))

# 3. è®©å®ƒåå¤æ€è€ƒè¿™ä¸ªè°œè¯­
for i in range(10):
    science_brain.teach("è°œè¯­", paradox)
    history.append(len(science_brain.stars))

# ç”»å›¾çœ‹çœ‹å‘ç”Ÿäº†ä»€ä¹ˆ
plt.figure(figsize=(10, 4))
plt.plot(history, linewidth=3, color='orange', marker='o')
plt.axvline(x=20, color='red', linestyle='--', label='Paradox!')
plt.title("Your Universe Just Leveled Up!", fontsize=15)
plt.xlabel("Time (Experience)", fontsize=12)
plt.ylabel("Intelligence (Number of Stars)", fontsize=12)
plt.grid(True, alpha=0.3)
plt.show()

print("ğŸ‰ çœ‹é‚£æ¡çº¢çº¿åé¢çš„è·³è·ƒï¼")
print("   é¢å¯¹å›°éš¾ï¼Œä½ çš„å®‡å®™æ²¡æœ‰å´©æºƒï¼Œè€Œæ˜¯ç¬é—´'å‡çº§'äº†ï¼")
print("   è¿™å°±æ˜¯ä¸ºä»€ä¹ˆæˆ‘ä»¬è¯´ï¼šå®‡å®™å¹¶éé™æ­¢çš„å­˜åœ¨ï¼Œè€Œæ˜¯æ°¸æ’ç”Ÿæˆçš„è¿‡åœºã€‚")
```
